```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
telco= pd.read_csv("telco.csv")
```

```python
telco.head()
```

```python
month_to_month= telco[telco.Contract == 'Month-to-month']
month_to_month.info()
```

```python
sns.relplot(x="tenure", y="MonthlyCharges", col= 'PaymentMethod',hue="Churn", data= month_to_month)
```

```python
sns.relplot(x="tenure", y="MonthlyCharges", col= 'PaymentMethod', style='SeniorCitizen', hue="Churn", data= month_to_month)
```

```python
sns.barplot(data=month_to_month, y='Churn', x='SeniorCitizen')
```

```python
senior_month=month_to_month[month_to_month.SeniorCitizen==1]
senior_month.head()
```

```python
sns.relplot(x="tenure", y="MonthlyCharges", col= 'PaymentMethod', hue="Churn", data= senior_month)
```

```python
month_churn=month_to_month['Churn'].value_counts()
senior_churned=senior_month['Churn'].value_counts()

print(month_churn, senior_churned)
```

```python
sns.barplot(x='PaymentMethod', y= senior_month['Churn'].value_counts(), data=senior_month)
```
